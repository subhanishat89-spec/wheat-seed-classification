To ensure valid paired statistical testing, prediction arrays were aligned to equal length before applying the paired t-test.

```python
# CELL 6 — Paired t-test

from scipy.stats import ttest_rel
import numpy as np

# Load correctness arrays
eff_correct = np.load(f"{BASE_PATH}/efficientnet_correct.npy")
res_correct = np.load(f"{BASE_PATH}/ResNet50_correct.npy")
vgg_correct = np.load(f"{BASE_PATH}/VGG16_correct.npy")

#  same length
min_len = min(len(eff_correct), len(res_correct), len(vgg_correct))

eff_correct = eff_correct[:min_len]
res_correct = res_correct[:min_len]
vgg_correct = vgg_correct[:min_len]

# Paired t-tests
t1, p1 = ttest_rel(eff_correct, res_correct)
t2, p2 = ttest_rel(eff_correct, vgg_correct)
t3, p3 = ttest_rel(res_correct, vgg_correct)

print("EfficientNet vs ResNet")
print("t-statistic:", t1)
print("p-value:", p1)

print("\nEfficientNet vs VGG")
print("t-statistic:", t2)
print("p-value:", p2)

print("\nResNet vs VGG")
print("t-statistic:", t3)
print("p-value:", p3)
```

f p < 0.05, the difference is statistically significant.
If p ≥ 0.05, the difference is not statistically significant.

```python
# CELL 7 — Interpretation of t-test

def interpret(p):
    if p < 0.05:
        return "Statistically Significant (p < 0.05) "
    else:
        return "Not Statistically Significant (p ≥ 0.05) "

print("Interpretation:\n")

print("EfficientNet vs ResNet:", interpret(p1))
print("EfficientNet vs VGG:", interpret(p2))
print("ResNet vs VGG:", interpret(p3))
```
